```python
data_root = '/content/dataset'
```

```python
import os
class_name = ['Golden_Retriever', 'Labrado_Retriever', 'Yorkshire_Terrier', 'Bulldog', 'Rottweiler', 'German_Shepherd', 'Dachshund', 'Poodle', 'Beagle', 'Boxer']

Golden_Retriever = data_root + '/Golden_Retriever'
Labrador_Retriever = data_root + '/Labrador_Retriever'
Yorkshire_Terrier = data_root + '/Yorkshire_Terrier'
Bulldog = data_root + '/Bulldog'
Rottweiler = data_root + '/Rottweiler'
German_Shepherd = data_root + '/German_Shepherd'
Dachshund = data_root + '/Dachshund'
Poodle = data_root + '/Poodle'
Beagle = data_root + '/Beagle'
Boxer = data_root + '/Boxer'

Golden_Retriever_name = os.listdir(Golden_Retriever)
Labrador_Retriever_name = os.listdir(Labrador_Retriever)
Yorkshire_Terrier_name = os.listdir(Yorkshire_Terrier)
Bulldog_name = os.listdir(Bulldog)
Rottweiler_name = os.listdir(Rottweiler)
German_Shepherd_name = os.listdir(German_Shepherd)
Dachshund_name = os.listdir(Dachshund)
Poodle_name = os.listdir(Poodle)
Beagle_name = os.listdir(Beagle)
Boxer_name = os.listdir(Boxer)

import PIL

x_train = []
y_train = []

for fname in Golden_Retriever_name:
  image = PIL.Image.open(Golden_Retriever+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(0)

for fname in Labrador_Retriever_name:
  image = PIL.Image.open(Labrador_Retriever+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(1)

for fname in Yorkshire_Terrier_name:
  image = PIL.Image.open(Yorkshire_Terrier+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(2)

for fname in Bulldog_name:
  image = PIL.Image.open(Bulldog+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(3)

for fname in Rottweiler_name:
  image = PIL.Image.open(Rottweiler+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(4)

for fname in German_Shepherd_name:
  image = PIL.Image.open(German_Shepherd+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(5)

for fname in Dachshund_name:
  image = PIL.Image.open(Dachshund+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(6)

for fname in Poodle_name:
  image = PIL.Image.open(Poodle+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(7)

for fname in Beagle_name:
  image = PIL.Image.open(Beagle+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(8)

for fname in Boxer_name:
  image = PIL.Image.open(Boxer+'/' + fname)
  image = image.resize((224,224))
  arr = np.array(image)
  x_train.append(arr)
  y_train.append(9)
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)

print(type(x_train), type(y_train))
print(x_train.shape, y_train.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, shuffle = True, random_state=42) # suffle은 기본값
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
np.random.seed(42)
sample = np.random.randint(773,size=25)
```

```python
fig = plt.figure(figsize = (8, 8))
for i, idx in enumerate(sample):
  plt.subplot(5, 5, i+1)
  plt.imshow(x_train[idx], cmap='gray')
  plt.axis('off')
  plt.title(class_name[y_train[idx]])
  fig.tight_layout()
plt.show()
```

## 전처리

```python
x_train_s = x_train.astype('float') / 255.
x_test_s = x_test.astype('float') / 255.
```

```python
from keras.utils import to_categorical

y_train_o = to_categorical(y_train)
y_test_o = to_categorical(y_test)
```

## 모델

```python
from keras import layers

model = keras.Sequential()

model.add(layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size = 2))
model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size = 2))
model.add(layers.Flatten())
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(10, activation = 'softmax'))
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
EPOCHS = 8
BATCH_SIZE = 32

history = model.fit(
    x_train_s, y_train_o,
    epochs = EPOCHS,
    batch_size = BATCH_SIZE,
    validation_data = (x_test_s, y_test_o)
)
```

## 평가

```python
y_pred = model.predict(x_test_s)
y_pred = np.argmax(y_pred, axis = 1)
y_pred
```

```python
def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch

    plt.figure(figsize=(16, 8))
    plt.subplot(1, 2, 1)
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.plot(hist['epoch'], hist['loss'], label='train loss')
    plt.plot(hist['epoch'], hist['val_loss'], label='val loss')
    plt.title('Loss Curve')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.xlabel('epochs')
    plt.ylabel('accuracy')
    plt.plot(hist['epoch'], hist['accuracy'], label='train accuracy')
    plt.plot(hist['epoch'], hist['val_accuracy'], label='val accuracy')
    plt.title('Accuracy Curve')
    plt.legend()
    plt.show()

plot_history(history)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score
from sklearn.metrics import confusion_matrix

def print_metrics(y_true, y_pred, aver='binary'):
    print('accuracy:', accuracy_score(y_true, y_pred))
    print('recall:', recall_score(y_true, y_pred ,average=aver))
    print('precision:', precision_score(y_true, y_pred,average=aver))
    print('f1 :', f1_score(y_true, y_pred, average=aver))

    cfm = confusion_matrix(y_true, y_pred)
    s = sns.heatmap(cfm, annot=True, cmap='Blues', fmt='d', cbar=False)
    s.set(xlabel='Prediction', ylabel='Actual')
    plt.show()
```

```python
print_metrics(y_test, y_pred, aver='macro')
```

## 전이학습

```python
from keras.applications import VGG16

base_model = VGG16(
    input_shape = (224, 224, 3),
    include_top = False,
    weights = 'imagenet'
)

base_model.trainable = False
base_model.summary()
```

```python
from keras import layers

model = keras.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dense(64, activation = 'relu'))
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(10, activation = 'softmax'))

model.summary()
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
EPOCHS2 = 15

history = model.fit(
    x_train_s, y_train_o,
    epochs = EPOCHS2,
    batch_size = BATCH_SIZE,
    validation_split=0.2
  )
```

```python
def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch

    plt.figure(figsize=(16, 8))
    plt.subplot(1, 2, 1)
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.plot(hist['epoch'], hist['loss'], label='train loss')
    plt.plot(hist['epoch'], hist['val_loss'], label='val loss')
    plt.title('Loss Curve')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.xlabel('epochs')
    plt.ylabel('accuracy')
    plt.plot(hist['epoch'], hist['accuracy'], label='train accuracy')
    plt.plot(hist['epoch'], hist['val_accuracy'], label='val accuracy')
    plt.title('Accuracy Curve')
    plt.legend()
    plt.show()

plot_history(history)
```
