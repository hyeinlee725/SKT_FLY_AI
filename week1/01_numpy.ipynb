## 인덱싱, 슬라이싱

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
arr[:] # 0~끝까지
arr[3:]  # 3~끝까지
arr[-1] # 맨뒤에 있는 것
arr[3:-1] # 3~끝에서 1개 앞까지
# arr[a:b] : a~(b-1)까지
arr[-3:] # 5, 6, 7
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
```

```python
arr
```

```python
arr[1, 1]
```

## 불리언 인덱싱

```python
arr = np.array([1, 2, 3, 4, 5])

# True인 것만 뽑아옴
idx = np.array([False, False, True, False, True])
arr[idx]
```

```python
# 3보다 큰 데이터만 읽어오기
arr > 3
```

```python
arr[arr > 3]
```

## fancy indexing

```python
# 내가 뽑아오고 싶은 자리의 위치 인덱스를 입력 -> 해당 위치인 인덱스 뽑아
arr = np.array([1, 2, 3, 4, 5])
idx = np.array([1, 3])
arr[idx]
```

```python
arr = np.array([1, 2, 3, 4, 5])
idx = np.array([[1, 3], [0, 1]])
arr[idx]
```

```python
arr = np.array([1, 2, 3, 4, 5])
arr[0] = -1
arr
```

## 배열의 형태 변환

```python
# 데이터 처리, 모델 학습 - 내가 만든 데이터 형태 2, 3차원, 모델이 받는 것이 4차원임
# 내가 가진 data 형태와 모델 input shape이 다른 경우가 많이 있음 -> 모델이 받는대로 바꿔줘야 함
# reshape
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr.shape)
# 2차원으로 변환
arr.reshape(3, 2)
```

```python
arr
```

```python
# 쭉 펴기
arr.reshape(-1)
arr.flatten()
arr.ravel()
```

```python
arr.reshape(1, 2, 3)
```

```python
arr.reshape(1, 2, -1) # -1자리는 알아서 계산해서 넣어달라는 의미
```

```python
arr.reshape(1, -1, 1)
```

```python
arr2 = arr.reshape(2, 3)
arr2
```

```python
np.sum(arr2)
```

```python
np.sum(arr2, axis=0)
```

```python
np.sum(arr2, axis=1)
```

```python
np.mean(arr2, axis=0)
```

```python
np.mean(arr2, axis=1)
```
