# Matplotlib
> 시각화 툴

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
matplotlib.__version__
```

```python
x = np.arange(0, 10, 0.1) # 0.1로 끊어서 나타내기
y = np.sin(x)
z = np.cos(x)
# 뒤에서 뭐가 일어났는지 모름
plt.figure(figsize=(4, 4)) #inch 단위
plt.plot(x, y, label="sin(x)", color="#ff0000")
plt.plot(x, z, label="cos(x)")
plt.legend()
plt.title("THIS IS LINE CHART")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# 최신 object 방법
x = np.arange(0, 10, 0.1) # 0.1로 끊어서 나타내기
y = np.sin(x)
z = np.cos(x)
# 객체 방식을 최근 많이 사용
# figure, axes를 불러옴
fig, ax = plt.subplots()
fig.set_size_inches(4, 4)
ax.plot(x, y, label="sin(x)", color="#ff0000")
ax.plot(x, z, label="cos(x)")
ax.set_title("THIS IS LINE CHART")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.legend()
```

## 산점도

```python
x = np.random.rand(30)
y = np.random.randint(0, 30, 30)
z = np.random.rand(0, 30, 30)

plt.scatter(x, y, marker='*')
plt.scatter(x, z, marker='*')
plt.show()
```

## 히스토그램

```python
# x축이 연속형 데이터임
x = np.random.randn(1000) # 정규분포에서 1000개를 뽑음
plt.hist(x, bins=30)
plt.show()
```

## Barplot

```python
x = ['A' , 'B', 'O', 'AB']
y = [10, 5, 6, 2]
plt.bar(x, y)
plt.show()
```

## Boxplot

```python
i1 = np.random.normal(0, 2.0, 1000)
i2 = np.random.normal(-3, 1.5, 500)
i3 = np.random.normal(3, 1.5, 500)
data = [i1, i2, i3]
plt.boxplot(data)
plt.show
```

## 여러개의 플롯 그리기

```python
x = np.linspace(0, 10)
y = np.sin(x)
z = np.cos(x)
w = np.tan(x)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r-')
plt.subplot(2, 2, 2)
plt.plot(x, z, 'g-')
plt.subplot(2, 2, 3)
plt.plot(x, w, 'b-')
plt.show
```
