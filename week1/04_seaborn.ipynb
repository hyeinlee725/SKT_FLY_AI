```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips, x = 'tip')
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='time', errorbar=None)
```

```python
sns.boxplot(data=tips, y='tip', hue='smoker')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='smoker', split=True)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2)
sns.boxplot(data=tips, y='tip', hue='smoker', ax=axes[0])
sns.violinplot(x='day', y='tip', data=tips, hue='smoker', split=True, ax=axes[1])

plt.show()
```
